```python
from collections import defaultdict
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

from data_utils import get_image_paths
```

```python
DATA_PATH = "../data/train"
```

```python
image_paths = get_image_paths(Path(DATA_PATH))
```

```python
image_samples = np.random.choice(image_paths, size=(5, 5), replace=False)

fig, ax = plt.subplots(5, 5, figsize=(12, 8))

for row_num, row in enumerate(image_samples):
    for col_num, image in enumerate(row): 
        img = Image.open(image["image"])
        ax[row_num, col_num].imshow(img)
        ax[row_num, col_num].axis("off")

plt.show()
```

```python
counter = defaultdict(int)

for image in image_paths:
    counter[image["label"]] += 1
```

```python

plt.title("Distribution of classes")
plt.xlabel("class num")
plt.ylabel("count")
plt.bar(counter.keys(), counter.values())
plt.show()
```
